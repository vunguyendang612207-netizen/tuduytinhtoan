Viết hàm kiểm tra một số có phải là số nguyên tố hay không? 


```python
def check_so_nguyen_to(n) :
    if n < 2 :
        return False
    for i in range(2, int(n**0.5) + 1 ) :
        if n % i == 0 :
            return False
    return True
print(check_so_nguyen_to(4))
```

## W5A4 
VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên. 


```python
def so_hoan_hao(n):
    s = 0
    for i in range(1, n) :
        if n % i == 0 :
            s += i
    return s == n
n = int(input())
print(so_hoan_hao(n))
```

## W5A8
Khoảng cách Hamming giữa hai số nguyên là số lượng vị trí khác nhau giữa hai dãy bits tương ứng
của chúng. Ví dụ: Khoảng cách Hamming giữa 1 (1) và 4 (100) là 2.
Cho hai số nguyên x và y, hãy viết hàm trả về khoảng cách Hamming giữa hai số này.
Khoảng cách Hamming là cái tên được đặt theo tên của Richard Hamming, người giới thiệu lý thuyết này trong tài liệu có tính cơ sở của ông về mã phát hiện lỗi và sửa lỗi (error-detecting and error-correcting codes). Nó được sử dụng trong kỹ thuật viễn thông để tính số lượng các bit trong một từ nhị phân (binary word) bị đổi ngược, như một hình thức để ước tính số lỗi xảy ra trong quá trình truyền thông, và vì thế, đôi khi, nó còn được gọi là khoảng cách tín hiệu (signal distance). Việc phân tích trọng số Hamming của các bit còn được sử dụng trong một số ngành, bao gồm lý thuyết tin học, lý thuyết mã hóa, và mật mã học. Tuy vậy, khi so sánh các dãy ký tự có chiều dài khác nhau, hay các dãy ký tự có xu hướng không chỉ bị thay thế không thôi, mà còn bị ảnh hưởng bởi dữ liệu bị lồng thêm vào, hoặc bị xóa đi, phương pháp đo lường phức tạp hơn, như khoảng cách Levenshtein (Levenshtein distance) là một phương pháp có tác dụng và thích hợp hơn

```python
def ten_to_bin(n: int) -> str:
    ans = ""
    while n > 0:
        du = n % 2
        ans = str(du) + ans
        n //= 2
    return ans
def hamming(n : int, m: int) -> int:
    n_bin = ten_to_bin(n)
    m_bin = ten_to_bin(m)
    while (len(n_bin) < len(m_bin)):
        n_bin = "0" + n_bin
    while (len(m_bin) < len(n_bin)):
        m_bin = "0" + m_bin
    ans = 0
    for i in range(len(n_bin)):
        if (m_bin[i] != n_bin[i]):
            ans += 1
    return ans

assert hamming(1, 4) == 2
assert hamming(123456, 32321) == 6
assert hamming(87654321, 21321835) == 15

print("All test completed")
```

```python
def tong_digit(n):
    tong = 0
    while n > 0:
        tong += n%10
        n //= 10
    return tong
assert tong_digit(103) == 4
assert tong_digit(251) == 8
assert tong_digit(12) == 3
assert tong_digit(3) == 3
assert tong_digit(1342) == 10
print("All tests completed")
```

## W5A12
Viết hàm divide_with_remainder(dividend, divisor) trả về bộ (q, r) sao cho dividend = divisor*q + r và
0 <= |r| < |divisor|. 

```python
def divide_with_remainder(a: int, b: int) -> tuple[int, int]:
    q = a//b
    r = a%b
    return q,r
assert divide_with_remainder(10, 2) == (5, 0)
assert divide_with_remainder(5, 2) == (2, 1)
assert divide_with_remainder(15, 4) == (3, 3)
assert divide_with_remainder(23, 2) == (11, 1)
assert divide_with_remainder(60, 5) == (12, 0)
print("All tests passed")
```

## W5A14 
Viết hàm compute(a, b, op) thực hiện phép tính a (+|-|*|/) b tùy theo ký tự op. 


```python
def compute(a: int, b: int, op: str) -> int:
    if op == "+":
        return a+b
    elif op == "-":
        return a-b
    elif op == "*":
        return a*b
    elif op == "/":
        return a/b
    else :
        return None
assert compute(4, 5, "+") == 9
assert compute(4, 5, "-") == -1
assert compute(9, 3, "*") == 27
assert compute(6, 3, "/") == 2
assert compute(5, 1, "%") == None
print("Completed all tests")
```

## W5A15 
Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông.

```python
def is_right_triangle(a: int, b: int, c: int) -> bool:
    if a**2 == b**2 + c**2:
        return True
    if b**2 == a**2 + c**2:
        return True
    if c**2 == a**2 + b**2:
        return True
    return False
assert is_right_triangle(3, 4, 5) == True
assert is_right_triangle(5, 12, 13) == True
assert is_right_triangle(7, 24, 25) == True
assert is_right_triangle(1, 2, 4) == False
assert is_right_triangle(5, 3, 9) == False
print("Completed all tests")
```
