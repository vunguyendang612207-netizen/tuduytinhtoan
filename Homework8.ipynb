## W8A16 
Yêu cầu
Xây dựng lớp Shape với phương thức area().
Xây dựng lớp Rectangle(w, h) và Circle(r) kế thừa từ Shape, ghi đè area().
Lấy π = 3.14.
Input
Dòng 1: loại hình (RECTANGLE hoặc CIRCLE)
Dòng 2: tham số tương ứng.
Output
In ra diện tích với 2 chữ số thập phân. 


```python
class Shape:
    def area(self):
        pass

class Rectangle(Shape):
    def __init__(self, w, h):
        self.w = w
        self.h = h
    def area(self):
        return self.w*self.h
class Circle(Shape):
    def __init__(self, r):
        self.r = r
    def area(self):
        return 3.14*(self.r**2)

sh = input()
if sh = "rectangle":
    w, h = map(float, input().split())
    hcn = Rectangle(w, h)
    print(f"{hcn.area():.2f}")
elif sh = "circle":
    r = float(input())
    htr = Circle(r)
    print(f"{htr.area():.2f}")
```

## W8A17 
Yêu cầu
Xây dựng lớp Vehicle(brand, speed).
Xây dựng lớp Car(brand, speed, seats) kế thừa từ Vehicle.
Viết phương thức info() cho mỗi lớp:
- Vehicle: brand-speed
- Car: brand-speed-seats
Input
Dòng 1: loại đối tượng (Vehicle hoặc Car)
Dòng 2: dữ liệu tương ứng.
Output
In ra chuỗi info(). 


```python
class Vehicle:
    def __init__(self, brand, speed):
        self.brand = brand
        self.speed = speed
    def info(self):
        return f"{self.brand}-{self.speed}"
class Car(Vehicle):
    def __init__(self, brand, speed, seats):
        self.brand = brand
        self.speed = speed
        self.seats = seats
    def info(self):
        return f"{self.brand}-{self.speed}-{self.seats}"
    
type = input()
data = input().split()
if type == "VEHICLE":
    brand = data[0]
    speed = data[1]
    veh = Vehicle(brand, speed)
    print(veh.info())
elif type == "CAR":
    brand = data[0]
    speed = data[1]
    seats = data[2]
    car = Car(brand, speed, seats)
    print(car.info())
    

    
```

## W8A18 Yêu cầu
Xây dựng lớp Polygon với phương thức perimeter().
Xây dựng lớp Triangle(a, b, c) kế thừa từ Polygon và ghi đè perimeter() trả về a+b+c.
Input
Một dòng chứa 3 số nguyên a b c.
Output
In ra chu vi tam giác.

```python
class Polygon:
    def perimeter(self):
        pass
class Triangle(Polygon):
    def __init__(self, a, b, c)
        self.a = a
        self.b = b
        self.c = c
    def perimeter(self):
        return self.a + self.b + self.c
a, b, c = map(int, input().split())
tamgi = Triangle(a, b, c)
print(tamgi.perimeter())
```

Xây dựng lớp Person(name, age) và lớp Student(name, age, school) kế thừa từ Person.
Phương thức describe():
- Person: name age
- Student: name age school
Input
Dòng 1: Person hoặc Student
Dòng 2: dữ liệu tương ứng.
Output
In ra kết quả describe().

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def describe(self):
        return f"{name} {age}"
class Student(Person):
    def __init__(self, name, age, school):
        self.name = name
        self.age = age
        self.school = school
    def  describe(self):
        return f"{name} {age} {school}"

p = input()
data = input().split()
if p == "Person":
    name = data[0]
    age = data[1]
    nguoi = Person(name, age)
    print(nguoi.describe())
elif p == "Student":
    name = data[0]
    age = data[1]
    school = data[2]
    hs = Student(name, age, school)
    print(hs.describe())
```

## W8A20 
Yêu cầu
Xây dựng lớp Account(balance) với hai phương thức deposit(x) và get_balance().
Xây dựng lớp SavingAccount(balance, rate) kế thừa từ Account, thêm phương thức
add_interest() để cộng balance*rate vào số dư.
Input
balance rate
Output
Số dư sau khi cộng lãi, in 2 chữ số thập phân. 


```python
class Account:
    def __init__(self, balance):
        self.balance = balance
    def deposit(self, x):
        self.balance += x
        return self.balance
    def get_balance(self):
        return self.balance
    
class SavingAccount(Account):
    def __init__(self, balance, rate):
        self.balance = balance
        self.rate = rate
    def add_interest(self):
        self.balance += self.balance*self.rate
nhap = (input().split())
balance = float(nhap[0])
rate = float(nhap[1])
NH = SavingAccount(balance, rate)
NH.add_interest()
print(f"{NH.get_balance():.2f}")


```

## W8A21 
Yêu cầu
Xây dựng lớp Printer với phương thức print_value(x).
Xây dựng lớp BinaryPrinter kế thừa từ Printer, ghi đè print_value(x) để in biểu diễn nhị phân của
số nguyên dương x. 
Input
Một số nguyên dương n.
Output
In ra dạng nhị phân của n. 

```python
class Printer:
    def print_value(self, x):
        print(f"{x}")
class BinaryPrinter(Printer):
    def print_value(self, x):
        if x == 0:
            print("0")
            return
        kq = ""
        while x > 0:
            kq = str(x % 2) + kq
            x //= 2
        print(kq)
n = int(input())
np = BinaryPrinter()
np.print_value(n)
            
```

## W8A22 
Yêu cầu
Xây dựng lớp Matrix2x2(a, b, c, d) biểu diễn ma trận:
[a b]
[c d]
Viết phương thức trace() trả về a + d và determinant() trả về ad - bc.
Input
Một dòng chứa 4 số nguyên a b c d.
Output
In ra trace và determinant trên cùng một dòng. 


```python
class Matrix2x2:
    def __init__(self, a, b, c, d):
        self.a = a
        self.b = b
        self.c = c
        self.d = d
    def trace(self):
        return a + d
    def determinant(self):
        return a*d - b*c

a, b, c, d = map(int, input().split())
Matran = Matrix2x2(a, b, c, d)
print(f"{Matran.trace} {Matran.determinant}")
```
