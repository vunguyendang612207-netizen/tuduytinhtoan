## Bài 1: 
Nhập vào 2 cạnh của một hình chữ nhật. In ra màn hình diệntíchvàchu vi của nó

Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểm tra xem nó có thể là độ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra màn hình.

```python
a, b, c = map(float, input().split())
if (a + b > c) and (a + c > b) and (b + c > a):
    if a == b == c:
        print("Day la tam giac deu")
    elif a == b or a == c or c == b:
        print("Day la tam giac can")
    elif a**2 == b**2 + c**2 or b**2 == a**2 + c**2 or c**2 == a**2 + b**2:
        print("Day la tam giac vuong")
    else:
        print("Day la tam giac thuong")
    chu_vi = a + b + c
    dien_tich = (chu_vi/2*(chu_vi/2-a)*(chu_vi/2-b)*(chu_vi/2-c))**0.5
    print(f'Dien tich cua tam giac la : {dien_tich}, Chu vi cua tam giac la {chu_vi}')
else:
    print("Khong la ba canh cua mot tam giac")
```

## Bài 5:
 Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
a, b, c = map(int, input().split())
denta = b**2 - 4*a*c
if a == 0 :
    print('Ko thoa man')
elif denta < 0 :
    print('Pt vo nghiem')
elif denta > 0 :
    print(f'Pt co nghiem la : x1 = {(-b - denta**0.5)/(2*a)} x2 = {(-b + denta**0.5)/(2*a)}')
else :
    print('Pt co nghiem kep x = {-b/(2*a)}')
```

Bài 9: Nhập vào thời gian 1 công việc nào đó là x giây. Hãy chuyển đổi và viết ra màn hình số thời gian trên dưới dạng bằng bao nhiêu giờ, bao nhiêu phút, bao nhiêu giây

```python
x = int(input())
gio = x // 3600 
phut = (x - gio*3600) // 60
giay = x - gio*3600 - phut*60
print(f'{gio} gio {phut} phut {giay} giay')
```

## Bài 10: 
Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vàomộtđiểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
a, b = map(int, input("Nhap vao tam : ").split())
r = int(input("Nhap vao ban kinh : "))
x, y = map(int, input("Nhap vao diem A :").split())
if (a - x)**2 + (b - y)**2 == r**2 :
    print("Diem A thuoc duong tron")
else :
    print("Diem A khong thuoc duong tron")
```

## Bài 11: 
Tính x^y (Với x, y là số thực). 

```python
x, y = map(float, input().split())
if y > 0 and y % 1 == 0 :
    print(x**y)
elif y < 0 and y % 1 == 0 and x != 0 :
    print(x**y)
elif y % 1 != 0 and x > 0 :
    print(x**y)
else :
    print("Khong hop le")
```
