## W8A1 
[Cylinder] 
Xây dựng lớp Cylinder (hình trụ) gồm: bán kính và chiều cao 
(kiểu số thực) với hai phương thức:tính diện tích bề mặt (diện 
tích toàn phần) và thể tích của hình trụ 
Đầu vào 
Đâu vào từ bàn phím gồm một dòng duy nhất chứa bán kính và
chiều cao của hình trụ, cách nhau bởi một dấu cách. 
Đầu ra
In ra màn hình diện tích bề mặt và thể tích của hình trụ (sử 
dụng lại những hàm vừa viết, π=3.14). 
(Kết quả làm tròn đến 2 chữ số sau dấu phẩy).

```python
class Cylinder:
    def __init__(self, r, h):
        self.r = r
        self.h = h
    def get_area(self):
        return 2*3.14*self.r**2 + 2*3.14*self.r*self.h
    def get_volume(self):
        return 3.14*self.r**2*self.h
    
a, b = map(int, input().split())
cy1 = Cylinder(a, b)
print(f'{cy1.get_area():.2f}, {cy1.get_volume():.2f}')
```

## W8A2 
[Date] 
Xây dựng lớp Date biểu diễn ngày tháng năm với hai phương 
thức kiểm tra xem một ngày tháng nào đó có hợp lệ hay không 
và phương thức tính ngày kế tiếp của ngày tháng năm nào đó 
nếu nó là hợp lệ 
Nhập vào một ngày tháng năm, (ngày tháng năm biểu diễn 
dưới dạng dd/mm/yyyy). Nếu ngày tháng năm không tồn tại, in 
ra dòng chữ INVALID. Nếu là hợp lệ in ra ngày kế tiếp của nó.
Đầu vào 
Đầu vào từ bàn phim chứa dữ liệu ngày, tháng, năm dưới dạng 
dd/mm/yyyy 
. 
Đầu ra
In ra màn hình ngày kế tiếp của ngày vừa nhập dưới dạng 
dd/mm/yyyy 
. Nếu ngày tháng không tồn tại, in ra INVALID

```python
class Date:
    def __init__(self, day, month, year):
        self.day = day
        self.month = month
        self.year = year
        self.max_day = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

    def leap_year(self):
        return (self.year % 400 == 0) or (self.year  % 4 == 0 and self.year % 100 == 0)
    
    def invalid_date(self):
        if self.leap_year():
            self.max_day[2] = 29
        if self.day > self.max_day[self.month]:
            return False
        return True
    
    def next_day(self):
        d = self.day
        m = self.month
        y = self.year 
        if d == 31 and m == 12:
            d = 1
            m = 1
            y += 1
        elif self.day == self.max_day[self.month]:
            d = 1
            m += 1
        else:
            d += 1
        return(f"{d}/{m}/{y}")

a, b, c = map(int, input().split("/"))
date = Date(a, b, c)

if date.invalid_date():
    print(date.next_day())
else:
    print("INVALID")
    


    
```

## W8A3 
[Point] 
Xây dựng lớp điểm với hai thuộc tính x, y cùng các phương 
thức kiểm tra xem có phải gốc tọa dộ không, có nằm trên trục 
tung không, có nằm trên trục hoành không, tính khoảng cách từ
điểm đó tới một điểm khác 
Đầu vào: 
Nhập vào tọa độ một điểm M bất kỳ 
Đầu ra:
Thông báo M có phải gốc tọa độ không hoặc có nằm trên trục 
tung hoặc trục hoành không, khoảng cách từ M tới gốc tọa độ 
là bao nhiêu (lấy 2 chữ số phần thập phân)

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def is_root_coordinates(self):
        if self.x == self.y == 0:
            return True
        return False
    
    def is_horizontal(self):
        if self.y == 0:
            return True
        return False
    
    def is_vertical(self):
        if self.x == 0:
            return True
        return False
    
    def distance(self):
        return (self.x**2+self.y**2)**0.5
    
x, y = map(int, input().split())
diem = Point(x, y)

if diem.is_root_coordinates():
    print('Diem M la goc toa do')
else:
    print('Diem M ko phai goc toa do')

if diem.is_horizontal():
    print('Diem M nam tren truc hoanh')
else:
    print('Diem M ko nam tren truc hoanh')

if diem.is_vertical():
    print('Diem M nam tren truc tung')
else:
    print('Diem M ko nam tren truc tung')

print(f'Khoang cach tu M den goc toa do la : {diem.distance():.2f}')
```

## W8A8 
Yêu cầu 
Lớp BankAccount(owner, balance) với: 
deposit(x), withdraw(x); nếu rút quá số dư thì bỏ qua giao dịch.
__str__() in owner: balance. 
Input 
Dòng 1: owner initial_balance 
Dòng 2: n (số lệnh) 
Tiếp theo n dòng, mỗi dòng: DEPOSIT x hoặc WITHDRAW x. 
Output 
Số dư cuối cùng (in số thực, 2 chữ số).

```python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = float(balance)
    def deposit(self, x):
        self.balance += float(x)
    def withdraw(self, x):
        x = float(x)
        if x <= self.balance:
            self.balance -= x
    def __str__(self):
        return f"{self.owner}: {self.balance}"

o, b = input().split()
account = BankAccount(o, b)
n = int(input())
for i in range(n):
    line = input().split()
    command = line[0]
    amount = line[1]
    if command == "DEPOSIT":
        account.deposit(amount)
    elif command == "WITHDRAW":
        account.withdraw(amount)
print(f"{account.balance:.2f}")


```

## W8A9 
Yêu cầu 
Lớp Student(name) với: 
add_score(subject, score) 
avg() trả về điểm trung bình. 
rank() trả về "Excellent" nếu avg ≥ 8, "Good" nếu 6.5–<8, 
"Average" nếu 5–<6.5, ngược lại "Poor". 
Input 
Dòng 1: name 
Dòng 2: n 
n dòng tiếp: subject score (score là số thực 0–10) 
Output 
name avg rank (avg 2 chữ số).

```python
class Student:
    def __init__(self, name):
        self.name = name
        self.scores = []
    def add_score(self, subject, score):
        self.scores.append(float(score))
    def avg(self):
        if not self.scores:
            return 0.0
        return sum(self.scores) / len(self.scores)
    def rank(self):
        average = self.avg()
        if average >= 8:
            return "Excellent"
        elif average >= 6.5:
            return "Good"
        elif average >= 5.5:
            return "Average"

        else:
            return "Poor"

name = input()
student = Student(name)
n = int(input())
for i in range(n):
    line = input().split()
    student.add_score(line[0], line[1])

print(f"{name} {student.avg():.2f} {student.rank()}")
```

## W8A10
Yêu cầu
Lớp Book(title, author, year).
Lớp Library() với:
add(book)
count_by_author(name)
find_by_year(y) trả về số sách xuất bản năm y.
Input
Dòng 1: n (số thao tác)
n dòng:
ADD title;author;year
COUNT author (đếm theo tác giả)
COUNTYEAR year (đếm theo năm)
Output
Mỗi lệnh COUNT/COUNTYEAR in một dòng kết quả

```python
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = int(year)

class Library:
    def __init__(self):
        self.books = []
    def add(self, book):
        self.books.append(book)
    def count_by_author(self, name):
        count = 0
        for book in self.books:
            if book.author == name:
                count += 1
        return count
    def find_by_year(self, y):
        count = 0
        for book in self.books:
            if book.year == int(y):
                count += 1
        return count
    
library = Library()
n = int(input())
for i in range(n):
    line = input().split(" ", 1)
    command = line[0]
    if command == "ADD":
        details = line[1].split(";")
        title = details[0]
        author = details[1]
        year = details[2]
        book = Book(title, author, year)
        library.add(book)
    elif command == "COUNT":
        author_name = line[1]
        print(library.count_by_author(author_name))
    elif command == "COUNTYEAR":
        year = line[1]
        print(library.find_by_year(year))




```
