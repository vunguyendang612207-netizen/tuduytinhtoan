## Bài 2:

```python
for i in range(1,100):
    if i % 3 == 0:
        print(i, end = ' ')
```

## Bài 3: 
Nhập vào số nguyên dương a, in ra bảng cửu chương của a

```python
a = int(input())
for i in range(1,11):
    print(f'{a} x {i} = {a*i}')
```

## Bài 5: 
Nhập vào số nguyên dương n. Tính S = 1 + 2 + 3 + 4 + … + n 

```python
n = int(input())
s = 0
for i in range(1,n+1):
    s = s + i
print(s)
```

## Bài 6: 
Nhập vào số nguyên dương a, in ra tất cả các ước dương của a

```python
a = int(input())
for i in range(1,a+1):
    if a % i == 0:
        print(i, end = ' ')
```

## Bài 7: 
Nhập vào số nguyên dương a, đếm số ước dương của a

```python
a = int(input())
dem = 0
for i in range(1,a+1):
    if a % i == 0:
        dem += 1
print(dem)
```

## Bài 8: 
Nhập vào số nguyên dương a và b, in toàn bộ ước dương chung của a và b 

```python
a, b = map(int, input().split())
so_nho = min(a,b)
for i in range(1, so_nho + 1):
    print(i)
```
