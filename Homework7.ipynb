## W7A16
Cho một mảng số nguyên không âm. Gọi mảng đó là "mảng đẹp" nếu
mỗi phần tử x trong mảng xuất hiện đúng x lần.
Kiểm tra xem mảng có phải là mảng đẹp hay không. Nếu không hãy xóa
đi y số (y là nhỏ nhất có thể) để mảng còn lại thành mảng đẹp.

```python
from collections import Counter

def check_beautiful_array(arr):
    counts = Counter(arr)
    total_removals = 0
    for k, v in counts.items():
        if v > k:
            total_removals += (v-k)
        elif v < k:
            total_removals += v
    return total_removals

ds = list(map(int, input().split()))
print(check_beautiful_array(ds))
    
```

## W7A18
Sắp xếp có điều kiện: Số chẵn đứng trước, số lẻ đứng sau, chẵn tăng
dần, lẻ giảm dần; 

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(len(arr)):
        min_id = i
        for j in range(i+1, len(arr)):
            if arr[min_id] > arr[j]:
                min_id = j
        arr[i], arr[min_id] = arr[min_id], arr[i]
    return arr

ds = list(map(int, input().split()))
ds_le = []
ds_chan = []
for n in ds:
    if n % 2 == 0:
        ds_chan.append(n)
    else:
        ds_le.append(n)
selection_sort(ds_le)
selection_sort(ds_chan)
print(ds_chan + ds_le[::-1])

```

## W7A20
Cho dãy N số nguyên dương, tính tổng các số có k chữ số. Dòng đầu ghi
N số nguyên dương, dòng sau ghi số nguyên dương k


```python
n = int(input())
ds = list(map(int, input().split()))
k = int(input())
def chu_so(x):
    x = abs(x)
    x = str(x)
    dem = 0
    for i in x:
        dem += 1
    return dem
tong = 0
for i in ds:
    if chu_so(i) == k:
        tong += i
print(tong)
```

## W7A22
Cho N tên sinh viên gồm cả họ và tên, tìm xem tên nào hay gặp nhất.
Nếu có nhiều tên cùng số lần gặp thì lấy tên có thứ tự từ điển nhỏ nhất

```python
from collections import Counter
n = int(input())
ds = []
for i in range(n):
    name = input().split()
    ds.append(name[-1])
lst = Counter(ds)
kq = min(lst.keys(), key=lambda x: (-lst[x], x))
print(kq)
```

## W7A26
Tách các số có trong một đoạn văn bản, tính tổng và tính trung bình các
số đó. 

```python
n = input()
ds = []
for i in n:
    if i.isnumeric():
        ds.append(int(i))
print(f"Tong : {sum(ds)} Trung binh: {sum(ds)/len(ds)}")
```
