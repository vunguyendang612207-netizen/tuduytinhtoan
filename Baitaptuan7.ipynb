## W7A1
Viết hàm tìm kiếm nhị phân
trên một danh sách đã
được sắp xếp tăng dần.
Trả về chỉ số của phần tử
hoặc -1 nếu không tìm
thấy

```python
def binary_search(arr, value):
    left, right = 0, len(arr) - 1
    while left <= right :
        mid = (left + right) // 2
        if arr[mid] == value:
            return mid
        elif arr[mid] < value:
            left = mid + 1
        elif arr[mid] > value:
            right = mid - 1
    return -1
lst1 = list(map(int, input().split()))
n = int(input())
print(binary_search(lst1, n))
```

## W7A2
Cho danh sách arr và một
số nguyên x, đếm xem x
xuất hiện bao nhiêu lần
trong danh sách.

```python
def count_occurences(arr, n) -> int:
    dem = 0
    for c in arr:
        if c == n:
            dem += 1
    return dem
arr = list(map(int, input().split()))
n = int(input())
print(count_occurences(arr, n))
```

## W7A9
Cho một danh sách các số 
nums, với mỗi phần tử 
nums[i], hãy đếm xem có 
bao nhiêu phần tử bên phải 
nhỏ hơn nums[i]. 
Trả về danh sách các số 
đếm. 
Efficient solution: using 
heap or binary search tree

```python
lst = list(map(int, input().split()))
res = []
for i in range(len(lst)):
    dem = 0
    for j in range(i, len(lst)):
        if lst[j] < lst[i]:
            dem += 1
    res.append(dem)
print(res)
```

nguyên nums và một số 
nguyên m, hãy tìm tổng 
mảng con lớn nhất theo 
modulo m. 
Tức là, với bất kỳ mảng 
con nums[i:j] nào, hãy tính 
sum(nums[i:j]) % m và trả 
về kết quả lớn nhất có thể.
Gợi ý: 
Duy trì tổng tiền tố (prefix 
sum) mod m và sử dụng 
tìm kiếm nhị phân để tìm 
tổng tiền tố nhỏ nhất chỉ 
lớn hơn tổng hiện tại một 
chút (để tối đa hóa hiệu số 
bao quanh).

```python
nums = list(map(int, input().split()))
m = int(input())
res = []
for i in range(len(nums)):
    for j in range(i+1, len(nums)):
        res.append(sum(nums[i:j])%m)
print(max(res))
```
