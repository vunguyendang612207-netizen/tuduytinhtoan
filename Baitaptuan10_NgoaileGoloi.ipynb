[DivideByZero]
Nhập vào hai số nguyên a, b. Thực hiện in ra kết quả phép chia a/b. Bắt lỗi ngoại lệ khi b

## W10A3
[ValueError - HappyNumbers]
Nhập vào một số nguyên dương a, Kiểm tra xem a có phải là số hạnh phúc không.
Một số nguyên dương n được gọi là số hạnh phúc nếu ta thực hiện liên tiếp các bước
sau:
- Tính tổng bình phương các chữ số của n.
- Lặp lại bước đó với kết quả vừa tính.
- Nếu cuối cùng kết quả là 1, thì n là số hạnh phúc.
- Nếu rơi vào vòng lặp vô hạn không bao giờ ra 1 (ví dụ như 4 → 16 → 37 → ...), thì n
không phải là số hạnh phúc.
Hãy nhập thử đầu vào là "a" rồi đọc kỹ Traceback để gỡ lỗi nhập liệu. (Dùng try ... except
bắt lỗi ValueError thêm set)

```python
try:
    a = int(input())
    seen = set()
    while a != 1 and a not in seen:
        seen.add(a)
        a = sum(int(digit)**2 for digit in str(a))
    if a == 1:
        print("True")
    else:
        print("False")
except ValueError:
    print("ValueError")
    
```

## W10A4
Sử dụng try except,kết hợp cùng else... finally.
Tính tổng 2 số, nếu nhập sai kiểu thì thông báo,
luôn luôn in "Kết thúc chương trình." sau khi chạy.


```python
try:
    a, b = map(int, input().split())
    kq = sum([a, b])
except ValueError:
    print("ValueError")
else:
    print(kq)
finally:
    print("Ket thuc chuong trinh")
```

## W10A12
[KeyError - StudentScore]
Cho trước từ điển điểm gồm các cặp khóa - giá trị: {"An": 8, "Binh": 7, "Chi": 9}. Nhập vào
tên học sinh và in ra điểm tương ứng. Nếu tên không tồn tại trong từ điển, bắt lỗi và in ra
"Khong tim thay hoc sinh".


```python
dic = {"An": 8, "Binh": 7, "Chi": 9}
try:
    name = input()
    check = True
    for k, v in dic.items():
        if k == name:
            check = False
            print(dic[k])
    if check == True:
        raise KeyError
except KeyError:
    print("Ko tim thay")
```

## W10A13
[IndexError - PopElement]
Nhập vào một danh sách số nguyên và một chỉ số k. Hãy in ra phần tử bị xóa khi thực
hiện pop(k). Nếu k không hợp lệ, in ra "Chi so khong hop le".

## W10A14
[TypeError - SumNumbers]
Nhập vào một danh sách gồm nhiều giá trị. Hãy cộng tất cả phần tử sau khi chuyển
được sang int. Nếu phần tử nào không chuyển được thì bỏ qua. In ra tổng cuối cùng. 

```python
ds = input().split()
s = 0
for c in ds:
    try:
        s += int(c)
    except ValueError:
        continue
print(s)
```

## W10A15
[Raise_Exception - PositiveNumber]
Nhập vào một số nguyên n. Nếu n <= 0 thì chủ động phát sinh ngoại lệ và in ra "So phai
duong". Nếu hợp lệ thì in ra bình phương của n

```python
try:
    n = int(input())
    if n <= 0:
        raise ValueError
    print(n**2)
except ValueError:
    print("So phai duong")
```
