## W4A19 [EvenNumBers]
Viết chương trình Python sử dụng một trong 2 loại vòng lặp for hoặc while để in ra các số
dương chẵn nhỏ hơn hoặc bằng n (n <= 1000) 


```python
n = int(input())
while n > 1000:
    n = int(input())
for i in range(1,n+1):
    if i % 2 == 0:
        print(i, end = " ")
```

## W4A21 
Nhập vào số nguyên dương n, tính tổng các chữ số của n

```python
n = int(input())
s = 0
while n > 0 :
    digit = n % 10
    s += digit
    n = n // 10
print(s)
```

## W4A22 
Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ

```python
n = int(input())
chan = 0
le = 0
while n > 0:
    digit = n % 10
    if digit % 2 == 0 :
        chan += 1
    else :
        le += 1
    n = n // 10
print(f'So chu so chan la : {chan} So chu so le la : {le}')
```

## W4A23

Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n 


```python
n = int(input())
s = 0
k = 1
while s + k < n :
    s += k
    k += 1
print(k-1)
```

## W4A24
Nhập vào A, tìm n nhỏ nhất sao cho 1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A

```python
a = int(input())
n = 1
s = 0
while ( s + 1/n <= a ) :
    s += 1/n
    n += 1
print(n)
```

## W4A25
Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1.
Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập 


```python
n = int(input())
max = n
min = n
while n != -1 :
    n = int(input())
    if n > max :
        max = n
    if n < min :
        min = n
print(max, min)
```

## W4A26 [Fibonaci1]
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ
bằng tổng hai số trước đó
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt
quá A

```python
n = int(input())
a = 0
b = 1
while b <= a:
    a, b = b, a + b
print(a)
```
