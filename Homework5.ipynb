## W5A16
Viết hàm kiểm tra một số có phải số Amstrong hay không? Hàm trả về kiểu bool, có tham số là một số
nguyên x

```python
def so_chu_so(n: int) -> int:
    dem = 0
    while n > 0:
        dem += 1
        n //= 10
    return dem

def check_amstrong(n: int) -> bool:
    so = n
    s = 0
    a = so_chu_so(n)
    while so > 0:
        digit = so%10
        s = s + digit**a
        so //= 10
    return s == n

assert check_amstrong(153) == True
assert check_amstrong(370) == True
assert check_amstrong(1) == True
assert check_amstrong(0) == True
assert check_amstrong(1634) == True
print("All tests completed")
```

## W5A18 
VIết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên.


```python
def doi_xung(n: int) -> bool:
    a = n
    s = 0
    while n > 0:
        digit = n%10
        s = s*10 + digit
        n //= 10
    return s == a
assert doi_xung(121) == True
assert doi_xung(123) == False
assert doi_xung(0) == True
assert doi_xung(24) == False
assert doi_xung(14541) == True
print("Passed all tests")

```

```python
def magic_number(n: int) -> bool:
    s = 0 
    while n > 0:
        digit = n%10
        s += digit
        n //= 10
    tong = 0
    while s > 0:
        so = s%10
        tong += so
        s //= 10
    return 1 == tong
assert magic_number(1234) == True
assert magic_number(31) == False
assert magic_number(5041) == True
assert magic_number(1) == True
assert magic_number(19) == True
print("All tests passed")
```

## W5A24 
Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không.


```python
def are_collinear(a: int, b: int, c: int, d: int, e: int, f: int) -> bool:
    x1 = c-a
    y1 = d-b
    x2 = c-e
    y2 = d-f
    return (x1*y2) == (x2*y1)
assert are_collinear(0, 0, 1, 1, 2, 2) == True
assert are_collinear(1, 5, 3, 5, 8, 5) == True
assert are_collinear(4, -2, 4, 0, 4, 5) == True
assert are_collinear(-2, -4, 0, 0, 1, 2) == True
assert are_collinear(1, 1, 1, 1, 5, 5) == True
assert are_collinear(0, 0, 0, 2, 2, 0) == False
assert are_collinear(1, 2, 3, 5, 4, 1) == False
assert are_collinear(0, 0, 1, 1, 2, 3) == False
print("All tests passed")
```

Viết hàm trả về số lượng chữ số khác 0 trong n số nguyên nhập vào từ bàn phím (một list)

```python
def count_0(l: list) -> int:
    dem = 0
    for i in l:
        for c in str(i):
            if c in "123456789":
                dem += 1
    return dem
assert count_0([0, 2, 4, 6, -9, -23]) == 6
assert count_0([-23, 34, 32]) == 6
assert count_0([234, 101, -100]) == 6
```

Viết hàm trả về giá trị lớn nhất và nhỏ nhất của ba số a,b,c dưới dạng list [max, min].

```python
def Max_min(a: int, b: int, c: int) -> list:
    lst = []
    if (a>=b) and (a>=c):
        lst.append(a)
    elif (b>=a) and (b>=c):
        lst.append(b)
    elif (c>=a) and (c>=b):
        lst.append(c)
    
    if (a<=b) and (a<=c):
        lst.append(a)
    elif (b<=a) and (b<=c):
        lst.append(b)
    elif (c<=a) and (c<=b):
        lst.append(c)
        return lst
    assert Max_min(5, 4, 1) == [5, 1]
    assert Max_min(-1, 2, 10) == [10, -1]
    assert Max_min(5, 5, 5) == [5, 5]
```

## W5A30 
Viết hàm tính hệ số nhị thức C(n, k).

```python
def giai_thua(n: int) -> int:
    gt = 1
    for i in range(1, n+1):
        gt = gt*i
    return gt
def he_so_nhi_thuc(n: int, k: int) -> int:
    if n < k:
        return None
    c = (giai_thua(n))/((giai_thua(k))*(giai_thua(n-k)))
    return c
assert he_so_nhi_thuc(5, 3) == 10
assert he_so_nhi_thuc(1, 3) == None
assert he_so_nhi_thuc(0, 0) == 1
```

## W5A34
Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau

```python
def chu_so_khac(n: int) -> bool:
    a = [0]*10
    while n>0:
        if a[n%10] == 1:
            return False
        a[n%10] = 1
        n //= 10
    return True
def sum_1_to_n(n: int) -> int:
    s = 0
    for c in range(1, n+1):
        if chu_so_khac(c):
            s += c
    return s
assert sum_1_to_n(10) == 55
assert sum_1_to_n(15) == 109
assert sum_1_to_n(25) == 292

                    
```
