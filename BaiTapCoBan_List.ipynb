### Bài 1: Tìm số lớn nhất của 1 list

```python
lst = list(map(int, input().split()))
maX = lst[0]
for i in lst:
    if i >= maX:
        maX = i
print(maX)
```

### Bài 3: Tìm giá trị trung bình của 1 list

```python
lst = list(map(int, input().split()))
s = 0
for x in lst:
    s += x
tb = s/len(lst)
print(f'{tb:.1f}')
```

### Bài 7: Kiểm tra một list số nguyên có phải là dãy tăng, dãy giảm hay vô danh
1 3 5 6 7 8 9: Tang
9 8 7 6 5 4 0: Giam
1 2 3 7 1 0 2: Vo danh
1 2 2 4 5 7: Vo danh

```python
lst = list(map(int, input().split()))
if lst[0] < lst[1]:
    check = 1
elif lst[0] > lst[1]:
    check = -1
else:
    check = 0
for i in range(1, len(lst)-1):
    if check == 1:
        if lst[i] <= lst[i-1]:
            check = 0
    if check == -1:
        if lst[i] >= lst[i-1]:
            check = 0
if check == 0:
    print("vo danh")
elif check == 1:
    print("tang")
else:
    print("giam")

```

### Bài 8: Kiểm tra một list số nguyên có phải là ziczac khong
1 3 2 5 2 8 7 9 2 100: ziczac
34 2 56 7 90 20 30 7 120: ziczac

```python
n = list(map(int, input().split()))
check = False
for i in range(1, len(n)-1):
    if (n[i-1] < n[i] > n[i+1]) or (n[i-1] > n[i] < n[i+1]):
        check = True
if check:
    print("ziczac")
else:
    print("not ziczac")
```

### Bài 9: Tìm và in ra tất cả các vị trí của số nguyên k trong một danh sách số nguyên cho trước. Nếu không có k trong ds thì in ra "-1".
VD: k = 3, [1,3,4,4,3,3,4,3,6,1] => k xuất hiện tại các vị trí (trả về 1 list) [1, 4, 5, 7]
k = 10, [1,3,4,4,3,3,4,3,6,1] => k không xuất hiện trong ds => trả về "-1"

```python
lst = list(map(int, input().split()))
k = int(input())
ls1 = []
for i in range(0, len(lst)):
    if lst[i] == k:
        ls1.append(i)
if len(ls1) == 0:
    print(-1)
else:
    print(ls1)


```

## Bài 3B
Tìm giá trị trung bình của các số nguyên dương từ danh sách số nguyên nhập vào từ bàn phím

```python
lst = list(map(int, input().split()))
ls = []
s = 0
for c in lst:
    if c > 0:
        ls.append(c)
for c in ls:
    s += c
print(s/len(ls))
```
