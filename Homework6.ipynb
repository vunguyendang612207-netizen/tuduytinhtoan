## W6A20
Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và
trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu không có thì trả về
-1


```python
def check(lst: list, n: int) -> int:
    for i in range(0, len(lst)):
        if lst[i] == n:
            return i+1
    return -1
lst = list(map(int, input().split()))
n = int(input())
print(check(lst, n))
```

## W6A23 
VIết hàm xóa đi các số <x có trong mảng

```python
def xoa_so(lst1: list, x: int) -> list:
    return [n for n in lst1 if n >= x]

liSt = list(map(int, input().split()))
x = int(input())
print(xoa_so(liSt, x))

```

## W6A24
Viêết hàm đẩy tất cả các số 0 về đầu (về cuối), các số khác vẫn đúng vị trí
trước sau: 0 2 3 2 4 0 0 0 9 => 0 0 0 2 3 2 4 9


```python
def ham_day(lst: list) -> list:
    return [n for n in lst if n != 0] + [c for c in lst if c == 0]
lst = list(map(int, input().split()))
print(ham_day(lst))
```

## W6A26
Tính tổng các phần tử cực đại địa phương trong một danh sách, không tính 2
phần tử đầu và cuối


```python
lst = list(map(int, input().split()))
s = 0
for i in range(1, len(lst)-1):
    if lst[i-1] <= lst[i] >= lst[i+1]:
        s += lst[i]
print(s)
```

## W6A27
Cho một dãy số nguyên, Tìm và in ra vị trí các số < x (x là giá trị cho trước).
nếu không có in ra -1


```python

lst = list(map(int, input().split()))
x = int(input())
res = [i for i, v in enumerate(lst, start=1) if v < x ]
if not res:
    print(-1)
else:
    print(res)
```
